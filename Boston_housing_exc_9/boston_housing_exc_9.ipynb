```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.datasets import boston_housing
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Lambda
from tensorflow.keras.utils import to_categorical
```

```python
(train_data, train_targest), (test_data, test_targets) = boston_housing.load_data()
```

```python
# РЎРѕР·РґР°РЅРёРµ DataFrame РґР»СЏ РїСЂРёР·РЅР°РєРѕРІ
feature_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 
    'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]
```

```python
df_train = pd.DataFrame(train_data, columns=feature_names[:-1])
df_train
```

```python
model = Sequential([
    Input(shape=(13,)),
    Dense(64, activation="relu"),
    Dense(1)
])
```

```python
model.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
```

```python
model.fit(train_data, train_targest, epochs=40, batch_size=16)
```
