```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model # шаблон линейной модели.
from sklearn.utils.validation import check_is_fitted
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_excel("gdprussia.xlsx")
df
```

```python
plt.scatter(df.oilprice, df.gdp)
plt.plot()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year','oilprice']], df[["gdp"]])
```

```python
check_is_fitted(reg, ["coef_", "intercept_"]) # Проверка, что модель была обучена с атрибутами "coef_" и "intercept_" Возвращает NONE если все хорошо
s = reg.predict(df[["year", "oilprice"]])
```

```python
plt.scatter(df.oilprice, df.gdp)
plt.plot(df.oilprice, s)
```
