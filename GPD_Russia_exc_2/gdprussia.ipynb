```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model # шаблон линейной модели.
```

```python
df = pd.read_excel("gdprussia.xlsx")
df
```

```python
plt.scatter(df.oilprice, df.gdp)
plt.plot()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["oilprice"]], df[["gdp"]])
```

```python
s = reg.predict(df[["oilprice"]])
```

```python
plt.scatter(df.oilprice, df.gdp)
plt.plot(df.oilprice, s)
```
