```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input, Lambda, Layer
```

```python
df = pd.read_csv("data.csv", names=['x', 'f(x)'])
df
```

```python
X = df["x"].values.astype(np.float32)
Y = df["f(x)"].values.astype(np.float32)

check_X = df["x"].isnull().sum() # РїСЂРѕРІРµСЂСЏРµРј РµСЃС‚СЊ Р»Рё РїСЂРѕРїСѓС‰РµРЅРЅС‹Рµ СЃС‚СЂРѕРєРё
check_Y = df["f(x)"].isnull().sum()
```

```python
class CustomLayer(Layer):
    def __init__(self, **kwargs):
        super(CustomLayer, self).__init__(**kwargs)
        self.a = self.add_weight(name='a', shape=(), initializer='ones')
        self.b = self.add_weight(name='b', shape=(), initializer='ones')
        self.c = self.add_weight(name='c', shape=(), initializer='ones')

    def call(self, inputs):
        x = inputs
        term1 = self.a * tf.sin(x)
        term2 = self.b * tf.math.log(x)
        term3 = (term1 + term2) ** 2
        term4 = self.c * x ** 2
        return term3 + term4
```

```python
input_layer = Input(shape=(1,))
output_layer = CustomLayer()(input_layer)
model = Model(inputs=input_layer, outputs=output_layer)
```

```python
model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(X, Y, epochs=1000)
```
