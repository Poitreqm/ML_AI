- We will solve the classification problem. We will predict which ofthe passengers survived the Titanic crash.
* Support Vector Machines

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, accuracy_score
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
columns_targed = ["Survived"]
columns_train = ["Pclass", "Sex", "Age", "Fare"]

# возьмем еще четыре колонки, которые, по нашему мнению, влияли на то, смогли спастись пассажир или нет. Эти колонки будут: пол пассажира, его возраст, класс каюты и тариф
# билета. Назовем их колонками для обучени
```

```python
X = df[columns_train]
Y = df[columns_targed]
```

```python
X["Sex"].isnull().sum()
```

```python
X["Pclass"].isnull().sum()
```

```python
X["Age"].isnull().sum() 
# Как вы понимаете, когда мы создаем
# модели, важны не только правильные алгоритмы, но и качество данных, на которых мы их
# обучаем. Имеется в виду, чтобы в наших данных не было пробелов или значений, которые
# там точно не могут быть. Например, при заполнении этой таблицы могли просто забыть вписать значение в какую-то ячейку, или вдруг значение возраста нечаянно перенесли в ячейку с
# тарифом. Поэтому в крупных проектах, когда дело идет на большие суммы, эти данные очень
# строго проверяют.
# Что же мы можем сделать чтобы проверить качество данных в данном проекте. Давайте
# как минимум проверим, что у нас нет пустых ячеек
# Вот нам выдало, что в колонке Age есть 177 пассажиров у которых пол не указан. Что
# делать?
```

```python
X["Fare"].isnull().sum()
```

```python
X.loc[:, "Age"] = X["Age"].fillna(X["Age"].mean())
```

```python
dict = {"male": 0, "female": 1}
X.loc[:, "Sex"] = X["Sex"].apply(lambda x: dict[x])
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
X_train
```

```python
model = svm.LinearSVC() # метод опорных векторов
```

```python
model.fit(X_train, np.ravel(Y_train)) # np.ravel преобразует двумерный массив в одномерный. хотя можно и без этого
```

```python
model.score(X_train, Y_train) # точность 78%
```

```python
predict_train = model.predict(X_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, Y_test) # нам показали 79% от всей выборки
```

```python
X.loc[:, 'Age'] = X['Age'].fillna(X['Age'].median())
```

```python
result = pd.DataFrame(predict_train).value_counts() # 0 - не выжили 
                                                    # 1 - выжили
result # данные которые я получил с помощью машинного обучения
```

```python
df["Survived"].value_counts() # изначальные данные 
```

```python
mse = mean_squared_error(Y_test, y_pred)
mse
```

```python
r2 = r2_score(Y_test, y_pred)
r2
```

```python
mae = mean_absolute_error(Y_test, y_pred)
mae
```

```python
accuracy = accuracy_score(Y_test, y_pred)
accuracy
```
