```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Flatten, Normalization
from tensorflow.keras.utils import to_categorical
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
wine = fetch_ucirepo(id=109) 

X = wine.data.features.values
y = wine.data.targets.values
```

```python
label_encoder = LabelEncoder()

y_encode = label_encoder.fit_transform(y) # РїСЂРµРѕР±СЂР°Р·РѕРІС‹РІРµРј РјРµС‚РєРё РёР· 1 2 
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y_encode, test_size=0.3, random_state=42)
```

```python
normalizer = Normalization() # РЅРѕСЂРјР°Р»РёР·Р°С†РёС†РёРѕРЅРЅС‹Р№ СЃР»РѕР№
normalizer.adapt(x_train)
```

```python
model = Sequential([
    Input(shape=(13,)),
    Dense(64, activation="relu"),
    Dense(64, activation="relu"),
    Dense(3, activation="softmax")
])
```

```python
model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',  # РїРѕС‚РѕРјСѓ С‡С‚Рѕ y_train СЌС‚Рѕ С†РµР»РѕС‡РёСЃР»РµРЅРЅС‹Рµ РјРµС‚РєРё РєР»Р°СЃСЃРѕРІ
              metrics=['acc'])
```

```python
model.fit(x_train, y_train, epochs=500, validation_data=(x_test, y_test))
```

```python
model.evaluate(x_test, y_test)
```

```python
pred = model.predict(x_train)
pred
```

```python
pred_class = np.argmax(pred, axis=1)
pd.DataFrame(pred_class).value_counts() * 100 / len(pred_class)
```
